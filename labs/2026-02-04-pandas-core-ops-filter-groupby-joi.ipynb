## Part 0 — Setup (Run This Cell)

This cell prepares SQLite, Pandas, and helper functions.


```python

import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")

def exec_sql(sql):
    conn.executescript(sql)
    conn.commit()

def q(sql):
    return pd.read_sql_query(sql, conn)

print("Environment ready.")
```


## Part 2 — Pull Data into a DataFrame

You will use **one SQL query** to create a DataFrame.
After this point, **do not use SQL** to answer questions.


```python

df = q('''
SELECT
    c.name AS customer,
    c.region,
    o.order_id,
    o.order_date,
    p.name AS product,
    p.category,
    oi.quantity,
    p.price,
    (oi.quantity * p.price) AS line_total
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
ORDER BY o.order_date;
''')

df
```

### Q5. Sort the DataFrame by line_total from highest to lowest.

### Q6. Create a new column called tax (7% of line_total).

```python

# TODO
df['tax'] = df['line_total'] * 0.07

df.head()

```

### Q7. Create a new column called total_with_tax.

```python

# TODO
df['total_with_tax'] = df['line_total'] + df['tax']

df.head()

```

### Q8. What is the total revenue per customer?

```python

# TODO
df.groupby('customer', as_index=False)['line_total'].sum().rename(columns={'line_total':'total_revenue'})

```

### Q9. What is the average line_total per product category?

```python

# TODO
df.groupby('category', as_index=False)['line_total'].mean().rename(columns={'line_total':'avg_line_total'})
```

### Q10. Which customer spent the most overall?

```python

# TODO
df.groupby('customer')['line_total'].sum().idxmax()

```

### Q11. Which region generated the most revenue?

```python

# TODO
df.groupby('region')['line_total'].sum().idxmax()

```

### Q12. Create a summary DataFrame with customer and total_spent.

```python

# TODO
customer_summary = (df.groupby('customer', as_index=False)['line_total']
                    .sum()
                    .rename(columns={'line_total':'total_spent'}))

customer_summary
```

### Q13. Write that summary DataFrame back to the database as customer_summary.

```python

# TODO
customer_summary.to_sql('customer_summary', conn, if_exists='replace', index=False)

q('SELECT * FROM customer_summary')
```
